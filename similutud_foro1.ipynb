#Ejemplo en Python: Manhattan

```python
import numpy as np
a = np.array([2, 3, 5])
b = np.array([7, 1, 9])
distancia = np.sum(np.abs(a - b))
print("Distancia Manhattan:", distancia)
# Resultado: 11
```

#Ejemplo en Python: Jaccard

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}
jaccard = len(set1.intersection(set2)) / len(set1.union(set2))
print("Similitud de Jaccard:", jaccard)
# Resultado: 0.428
```

```python
import numpy as np
from sklearn.metrics import jaccard_score

# predicted values
y_pred = np.array([1, 1, 1, 0, 1]).reshape(-1, 1)
# true values
y_true = np.array([1, 1, 0, 0, 1]).reshape(-1, 1)

# Calculate Jaccard Index
jaccard_index = jaccard_score(y_true, y_pred)

# Calculate Jaccard Distance
jaccard_distance = 1 - jaccard_index

print("Jaccard Index:", jaccard_index)
print("Jaccard Distance:", jaccard_distance)
```
